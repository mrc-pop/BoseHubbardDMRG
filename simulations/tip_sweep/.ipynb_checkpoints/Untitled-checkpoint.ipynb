```julia
using DelimitedFiles
using Plots; pgfplotsx()
using LaTeXStrings
using LsqFit
using ColorSchemes
```

```julia
L = 50
Data = readdlm("L=50_site=25.txt", ';', '\n'; comments=true);
```

```julia
# Extract unique J, μ values
JJ = unique(Data[:, 1])
μμ = unique(Data[:, 2])

# Mastruzzo to extract array of C, eC
function parse_array(str)
    return parse.(Float64, split(strip(str, ['[', ']', ' ']), ','))
end
Call = [parse_array(row[8]) for row in eachrow(Data)];
eCall = [parse_array(row[9]) for row in eachrow(Data)];
```

```julia
p = plot()
MyColors = ColorSchemes.tab10

for (j,J) in enumerate(JJ)
    # println("Chosen J=$J")
    
    # Filter data for the current J
    filter = (Data[:, 1] .== J)

    C = Call[filter][1]
    r = 1:1:length(C)
        
    J_round = round(J, digits=3)

    scatter!(p, r, -C,
        xlabel=L"$r$",
        ylabel=L"$|C(r)|$",
        title=L"Density-density correlation function ($L=%$L$)",
        label=L"J=%$J_round",
        markersize=2,
        legend=:topright,
        #yscale=:log10,
        xscale=:log10,
        color=MyColors[j],
        ylimit=(0.0, maximum(-C))
        )
    
    plot!(p, r, -C,
        label=:none,
        alpha=0.5,
        markersize=2,
        color=MyColors[j],
        ylimit=(0.0, maximum(-C))
        )
    
end
p
gui()
```

```julia
r_min_fit = 2
r_max_fit = 20

C_cut = abs.(C[r_min_fit:r_max_fit])

# Define the fit function
fit_func(x, p) = p[1]*x.^(-p[2])  # power-law
p0 = [1.0, 1.0]

fit = curve_fit(fit_func, r_min_fit:r_max_fit, C_cut, p0) 

exponent = fit.param[2]
e_exponent = stderror(fit)[2]

print("Exponent: $exponent +/- $e_exponent")

fit_x = range(r_min_fit, r_max_fit, length=80)
fit_y = fit_func(fit_x, fit.param)

plot!(p, fit_x, fit_y, label=false, alpha=0.7)
```
